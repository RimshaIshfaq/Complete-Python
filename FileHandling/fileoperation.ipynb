## File Operation- Read And Write Files
File handling is a crucial part of any programming language. Python provides built-in functions and methods to read from and write to files, both text and binary. This lesson will cover the basics of file handling, including reading and writing text files and binary files.

```python
## read a whole file
with open('example.txt','r') as file:
    data = file.read()
    print(data)
```

```python
## read  a file line by line

with open('example.txt','r') as file:
    for line in file:
        print(line.strip())  # strip() is used to remove the newline character at the end of each line
```

```python
## open a file for writing something in it (overwriting)

with open('example.txt', 'w') as file:
    file.write('This is a new line.\n')
    file.write('Hi guys, I love you all.\n')
```

```python
## Write a file without overwriting (append mode)

with open('example.txt', 'a') as file:
    file.write('This is an appended line.\n')
    file.write('I am appending this line to the file.\n')
```

```python
## Writing a list of lines to a file

lines = ['Line 1\n', 'Line 2\n', 'Line 3\n']

with open('example.txt', 'a') as file:
    file.writelines(lines)
```

```python
### Binary Files
# in binary files,we usually update bytes

data = b'\x00\x02\x04\x06\x08'
with open('example.bin', 'wb') as file:
    file.write(data)
```

```python
# read  a file line by line
with open('example.bin','rb') as file:
    content = file.read()
    print(content)
```

```python
# read the content from source text file and write it to a destination text file

with open('example.txt', 'r') as source_file:
    content = source_file.read()

with open('destination.txt', 'w') as dest_file:
    dest_file.write(content)
    
```

```python
# read a text file and count the number of lines, words and characters in it

def count_file_content(file_path):
    with open(file_path,'r') as file:
        lines = file.readlines()
        num_lines = len(lines)
        num_words = sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)
    return num_lines, num_words, char_count

file_path = 'example.txt'
lines, words, characters = count_file_content(file_path)
print(f'Lines: {lines}, Words: {words}, Characters: {characters}')
```

```python
# writing and then reading a file

with open('example.txt', 'w+') as file:  # 'w+' mode allows us to write and then read from the same file
    file.write('Hello world\n')
    file.write('Welcome to file handling in Python.\n')

    file.seek(0) # Move the file cursor back to the beginning of the file

    content = file.read()
    print(content)
```
