```python
# if statement 
# it evaluates teh situation and executes the block of code if situation is true

age = 24

if age >=18:
    print("You are eligible to vote")
```

```python
# else statement
# it executes the block of code if the condition in if statement is false

age = 16

if age >=18:
    print("You are eligible to vote")
else:
    print("You are not eligible to vote")
```

```python
# elif statement
# it is used to check multiple conditions. It stands for else if

age = 70
if age < 18:
    print("You are a child")
elif age < 60:
    print("You are an adult")
else:
    print("You are a senior citizen")
```

```python
## Nested Conditional statements

num = 34

if num > 0:
    if num % 2 == 0:
        print("The number is positive and even")
    else:
        print("The number is positive and odd")
else:
    print("The number is not positive")
    
```

```python
## Practical Examples

## Determine if a year is a leap year using nested condition statement

year=int(input("Enter the year"))

if year%4==0:
    if year%100==0:
        if year%400==0:
            print(year,"is a leap year")
        else:
            print(year,"is not a leap year")
    else:
        print(year,"is a leap year")

else:
    print(year,"is not a leap year")
```

```python
# Assignment
# Simple Calculator using if-else statements

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operation = input("Enter operation (+, -, *, /): ")

if operation == '+':
    result = num1 + num2
elif operation == '-':
    result = num1 - num2
elif operation == '*':
    result = num1 * num2
elif operation == '/':
    if num2 != 0:
        result = num1 / num2
    else:
        print("Error: Division by zero is not allowed.")
else:
    print("Invalid operation. Please enter one of +, -, *, /.")

print("Result:", result)
```

```python
### Determine the ticket price based on age and whether the person is a student.
# Ticket pricing based on age and student status

# Take user input

age = int(input("Enter your age: "))
is_student = input("Are you a student? (yes/no): ").strip().lower()

# Determine ticket price

if age < 12:
    ticket_price = 5  # Child ticket price
elif age < 18:
    if is_student == 'yes':
        ticket_price = 7  # Student ticket price for teenagers
    else:
        ticket_price = 10  # Regular ticket price for teenagers
elif age < 65:
    if is_student == 'yes':
        ticket_price = 8  # Student ticket price for adults
    else:
        ticket_price = 15  # Regular ticket price for adults
else:
    ticket_price = 10  # Senior citizen ticket price

print("Your ticket price is: $", ticket_price)
```

```python
## Employee Bonus Calculation
# Calculate an employee's bonus based on their performance rating and years of service.

years_of_service = int(input("Enter years of service: "))
performance_rating = input("Enter performance rating (excellent, good, average, poor): ").strip().lower()

# Determine bonus percentage based on performance rating

if performance_rating == 'excellent':
    if years_of_service >= 5:
        bonus_percentage = 20
    else:
        bonus_percentage = 15
elif performance_rating == 'good':
    if years_of_service >= 5:
        bonus_percentage = 10
    else:
        bonus_percentage = 5
elif performance_rating == 'average':
    if years_of_service >= 5:
        bonus_percentage = 5
    else:
        bonus_percentage = 2
elif performance_rating == 'poor':
    bonus_percentage = 0
else:
    print("Invalid performance rating. Please enter one of excellent, good, average, poor.")
    bonus_percentage = 0

print("Your bonus percentage is:", bonus_percentage, "%")
      
```

```python
## User Login System
# A simple user login system that checks the username and password.

# Predefined username and password

predefined_username = "admin"
predefined_password = "password123"

# Take user input for username and password

username = input("Enter username: ")
password = input("Enter password: ")

# Check if the username and password are correct

if username == predefined_username:
    if password == predefined_password:
        print("Login Successful!")
    else:
        print("Incorrect password. Please try again.")
else:
    print("Username not found. Please try again.")
```
