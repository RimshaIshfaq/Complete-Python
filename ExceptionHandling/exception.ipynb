## Exception Handling

#### Understanding Exceptions
Exception handling in Python allows you to handle errors gracefully and take corrective actions without stopping the execution of the program. This lesson will cover the basics of exceptions, including how to use try, except, else, and finally blocks.

#### What Are Exceptions?
Exceptions are events that disrupt the normal flow of a program. They occur when an error is encountered during program execution. Common exceptions include:

- ZeroDivisionError: Dividing by zero.
- FileNotFoundError: File not found.
- ValueError: Invalid value.
- TypeError: Invalid type.

```python
a=b #Name Error Exception is raised
```

```python
# to handle it , I will use try and except block
try:
    a=b
except:
    print("The variable has not been assigned")
```

```python
# or 
try:
    a=b
except NameError as ex:
    print(ex)
```

```python
result = 1/0 #will get zero division error
```

```python
# let's handle it
try:
    result = 1/0
except ZeroDivisionError as ex:
    print(ex)
    print("please enter denominator greater than zero")
```

```python
try:
    result = 1/2
    a=b
except ZeroDivisionError as ex:
    print(ex)
    print("please enter denominator greater than zero")
```

```python
## every exception is derived from the parent class which is "Exception", so in the above case, we will use this
# let's handle it
try:
    result = 1/2
    a=b
except ZeroDivisionError as ex:
    print(ex)
    print("please enter denominator greater than zero")
except Exception as ex1: #we can use any no. of exception as we want just make sure the last one is "Exception"
    print(ex1)
    print("Main exception got caught here")
```

```python
try:
    num = int(input("Enter a valid number:"))
    result = 10/num
except ValueError:
    print("This is not a valid number")
except ZeroDivisionError:
    print("Enter number greater than zero")
except Exception as e:
    print(e)
```

```python
# try, except, else block
try:
    num = int(input("Enter a valid number"))
    result = 10/num
except ValueError:
    print("This is not a valid number")
except ZeroDivisionError:
    print("Enter number greater than zero")
except Exception as e:
    print(e)
else:
    print(f"the result is {result}")
```

```python
# try, except, else and finally 
try:
    num = int(input("Enter a valid number"))
    result = 10/num
except ValueError:
    print("This is not a valid number")
except ZeroDivisionError:
    print("Enter number greater than zero")
except Exception as e:
    print(e)
else:
    print(f"the result is {result}") # only when there is no error
finally:
    print("Execution Completed")
```
