#### Introduction to Functions
#### Definition:

A function is a block of code that performs a specific task. Functions help in organizing code, reusing code, and improving readability.

```python
## syntax
def function_name(parameters):
    """Docstring"""
    # Function body
    return expression
```

```python
## why functions

num = 33
if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
def even_odd(num):
    """Returns whether a number is even or odd"""
    if num % 2 == 0:
        return "Even"
    else:
        return "Odd"
```

```python
even_odd(44)
```

```python
even_odd(89)
```

```python
# function with multiple parameters
def add_numbers(a, b):
    """Returns the sum of two numbers"""
    return a + b

add_numbers(5, 7)
```

```python
## Default Parameters

def greet(name):
    print(f"Hello {name} Welcome To the paradise")

greet("Rimsha")
```

```python
### Variable Length Arguments
## Positional And Keywords arguments

def print_numbers(*args):
    """Prints all the numbers passed as arguments"""
    for num in args:
        print(num)
```

```python
print_numbers(1,2,3,4,5)
```

```python
### Keywords Arguments

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(name="Rimsha",age=25, city="Lahore")
```

```python
def print_details(*args, **kwargs):
    for val in args:
        print(f" Positional argument: {val}")
    
    for key, value in kwargs.items():
        print(f"Keyword argument: {key} = {value}")
```

```python
print_details(1,2,3,4,5,"Rimsha", name="Rimsha", age=24, city="Lahore")
```

```python
def multiply(a,b):
    return a * b

multiply(2,3)
```

```python
# Return multiple parameters

def multiply(a,b):
    return a*b, b


multiply(2,3)
```
