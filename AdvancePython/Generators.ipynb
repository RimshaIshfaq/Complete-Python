## Generators

Generators are a simpler way to create iterators. They use the yield keyword to produce a series of values lazily, which means they generate values on the fly and do not store them in memory.

They are subclass of iterators.

```python
def square(n):
    for i in range(3):
        return i**2
```

```python
square(3)
```

```python
def square(n):
    for i in range(3):
        yield i**2
```

```python
square(3)
```

```python
for i in square(3):
    print(i)
```

```python
a = square(3)
a
```

```python
next(a)
```

```python
def my_generator():
    yield 1
    yield 2
    yield 3
```

```python
gen = my_generator()
gen
```

```python
next(gen)
```

```python
for val in gen:
    print(val)
```

### Practical example is : Reading Large Files

Generators are particularly useful when dealing with large datasets or files that cannot fit into memory. Instead of loading the entire file into memory, you can read it line by line using a generator.

```python
## Practical: Reading Large Files

def read_large_file(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line
```

```python
file_path = 'large_file.txt'

for line in read_large_file(file_path):
    print(line.strip())
```
