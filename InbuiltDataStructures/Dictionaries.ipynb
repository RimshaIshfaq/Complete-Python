#### Introduction to Dictionaries
Dictionaries are unordered collections of items. They store data in key-value pairs. Keys must be unique and immutable (e.g., strings, numbers, or tuples), while values can be of any type.

```python
# creating dictionaries
empty_dict ={}
print(empty_dict)
print(type(empty_dict))
```

```python
my_dict = dict()
print(my_dict)
print(type(my_dict))
```

```python
student = {"name":"Rimsha", "age":23, "courses":["Data Science","Machine Learning"]}
print(student)
```

```python
#Accessing elements
print(student["name"])

#Accessing using get method
print(student.get("age"))
print(student.get("last_name"))
```

```python
# only single key is used
biodata = {"name":"Rimsha", "age":23, "name":43}
biodata
```

```python
## Modifying Dicitonary Elements
## Dictionary are mutable,so you can add, update or delete elements
print(student)
```

```python
student["age"] = 24
print(student)

student["address"] = "Pakistan"
print(student)
```

```python
del student['courses'] ## delete key and value pair

print(student)
```

```python
## Dictionary Methods

keys = student.keys()
print(keys)

values = student.values()
print(values)

items = student.items()
print(items)
```

```python
# shallow copy
student_copy = student
print(student_copy)
print(student)
```

```python
student["name"] = "Meow"
print(student_copy)
print(student)
```

```python
student_copy1 = student.copy() ## deep copy
student["name"] = "Engr Rimsha"
print(student_copy1)
print(student)
```

```python
### Iterating Over Dictionaries
## You can use loops to iterate over dictionatries, keys,values,or items

## Iterating over keys
for key in student.keys():
    print(key)
```

```python
# iterating over values
for value in student.values():
    print(value)
```

```python
# iterating over items
for item in student.items():
    print(item)
```

```python
## Nested Disctionaries
students={
    "student1":{"name":"Akash","age":32},
    "student2":{"name":"Peter","age":35}
}
print(students)
```

```python
## Access nested dictionaries elementss
print(students["student2"]["name"])
print(students["student2"]["age"])
```

```python
students.items()
```

```python
# Iterating over nested dictionaries
for student_id, student_info in students.items():
    print(f"Student ID: {student_id}")
    print(f"Name: {student_info['name']}")
    print(f"Age: {student_info['age']}")
    print()
```

```python
# Dictionary Comprehension

squares = {x: x**2 for x in range(1,6)}
print(squares)
```

```python
## Practical Examples

## USe a dictionary to count he frequency of elements in list

numbers = [1,2,2,3,3,3,4,4,4,4]
frequency = {}

for number in numbers:
    if number in frequency:
        frequency[number] += 1
    else:
        frequency[number] = 1

print(frequency)
```

```python
## Merge 2 dictionaries into one

dict1 = {"a":1, "b":2}
dict2 = {"c":3, "d":4}
merged_dict = {**dict1, **dict2}
print(merged_dict)
```
